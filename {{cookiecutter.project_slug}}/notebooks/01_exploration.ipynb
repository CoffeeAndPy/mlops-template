## 3. Null values

Before carrying out any analysis, it's crucial to check for null values in the dataset. Null values can lead to errors in modeling and can skew the results if not handled properly.

````python
nulls = df.isnull().sum()
nulls_pct = (nulls / len(df) * 100).round(2)

null_report = pd.DataFrame({'count': nulls, 'pct': nulls_pct})
null_report[null_report['count'] > 0].sort_values('pct', ascending=False)
````
